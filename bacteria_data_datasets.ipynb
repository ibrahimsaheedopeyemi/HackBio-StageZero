# Task Description: Data Exploration and Cleaning with Pandas

```python
#Part 1: Basic Data Inspection
import pandas as pd
url = "https://raw.githubusercontent.com/HackBio-Internship/2025_project_collection/refs/heads/main/bacteria.csv"
bacteria_data = pd.read_csv(url)
```

```python
# Shape of the dataset
print(bacteria_data.shape)
```

```python
#  List of all Column names
print(bacteria_data.columns.tolist())
```

## Part 2: Data Selection


```python
# Select row 15
print(bacteria_data.iloc[[15]])
```

```python
# Select the column species
print(bacteria_data["species"])
```

```python
# Retrieve the value at row 15 for the column Isolation Origin.
print(bacteria_data.loc[15, "Isolation origin"])
```

## Part 3: Data Filtering

```python
# Filter rows where carb_fit is greater than 0.5
filtered = bacteria_data[bacteria_data["carb_fit"] > 0.5]
print(filtered)
```

## Part 4: Handling Missing Data


```python
# Create a copy
bacteria_data_missing = bacteria_data.copy()
```

```python
# Count missing values per row
print(bacteria_data_missing.isna().sum())
```

```python
# Fill missing values with one
filled_bacteria_data = bacteria_data_missing.fillna(1)
print(filled_bacteria_data)
```

```python
# Drop columns with missing values
dropped_bacteria_data = bacteria_data_missing.dropna(axis = 1)
print(dropped_bacteria_data)
```

## Part 5: Checking Missing Data in Main Dataset


```python
# Check if missing values exist
print(bacteria_data.isnull().any())
```

```python
# Count number of missing values per column
print(bacteria_data.isna().sum())
```

## Part 6: Column Renaming


```python
# Rename columns `C1' and `C2`
bacteria_data = bacteria_data.rename(columns = {"C1": "UMAO1", "C2": "UMAP2"})
```

```python
bacteria_data.columns
```

## Part 7: Sorting

```python
sorted_data = bacteria_data.sort_values(by = "BSL", ascending = False)
print(sorted_data)
```

# Part 8: Grouped Analysis


```python
# Compute mean grouped of cipro_fit and carb_fit by label type
print(bacteria_data.groupby("labels")[["cipro_fit", "carb_fit"]].mean())
```

```python
# Compute max grouped of cipro_fit and carb_fit by label type
print(bacteria_data.groupby("labels")[["cipro_fit", "carb_fit"]].max())
```
